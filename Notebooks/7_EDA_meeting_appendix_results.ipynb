```python
def classify_location(location):
    if pd.isna(location):
        return None
    p = location.lower()
    if "mixed-use" in p:
        return "Mixed-Use"
    elif ("residence" in p or "residential" in p) and ("commercial" in p or "office" in p or "retail" in p or "restaurant" in p):
        return "Mixed-Use"
    elif ("medstar" in p and "georgetown" in p) or ("medstar" in p and "hospital" in p):
        return "Medstar Georgetown Hospital"
    elif "georgetown university" in p:
        return "Georgetown University"
    elif "residence" in p or "residential" in p:
        return "Residential"
    elif "office" in p:
        return "Office"
    elif "commercial" in p or "retail" in p or "restaurant" in p or "café" in p or "cafe" in p:
        return "Commercial"
    # elif "alterations to" in p:
    #     return "Alterations"
    elif "church" in p or "congregation" in p:
        return "Church/Congregation"
    elif "dumbarton oaks" in p or "tudor place" in p:
        return "Dumbarton Oaks/Tudor Place"
    elif "hotel" in p:
        return "Hotel"
    elif "condominium" in p:
        return "Condominiums"
    elif "bank" in p:
        return "Bank"
    elif "school" in p:
        return "School"
    elif " house" in p:
        return "House"
    else:
        return "Other/Unclassified"
```

```python
cfa['property'] = cfa['property'].str.strip()
```

```python
cfa['property_categorized'] = cfa['property'].apply(classify_location)
```

```python
# This is crazy the number of residential properties as a share of all the projects.
cfa['property_categorized'].value_counts()
```

```python
# Over half of the projects contain the labels "Residential" or "Residence"
cfa['property_categorized'].value_counts(normalize=True)
```

review type

```python
def classify_review_type(review_type):
    if pd.isna(review_type):
        return None
    p = review_type.lower()
    if "concept" in p:
        return "Concept"
    elif "permit" in p:
        return "Permit"
    elif "informational" in p:
        return "Informational"
    else:
        return "Unknown"
```

```python
# cfa['review_type'].apply(classify_review_type).value_counts()
cfa['review_type_categorized'] = cfa['review_type'].apply(classify_review_type)
```

```python
# cfa[cfa['review_type_categorized'] == "Unknown"]
cfa.loc[(cfa['review_type_categorized'] == "Unknown"),'review_type_categorized'] = cfa.loc[(cfa['review_type_categorized'] == "Unknown"),'recommendation'].apply(classify_review_type) 
```

### Combining Consent and Denial

```python
decisions = pd.concat([consent,denial])
```

```python
decisions['STATUS'].value_counts(normalize=True)
```

```python
def get_recommendation_categorized(rec):
    if pd.isna(rec):
        return "Unknown"
    rec = rec.strip().lower()
    if 'submitted materials are not sufficient' in rec or 'submitted material is not sufficient' in rec or "insufficient documentation" in rec or "materials are insufficient" in rec or "insufficient information" in rec or "documentation is not sufficient" in rec:
        return "Insufficient Materials"
    elif "incompatible with the historic district" in rec or "incompatible with historic district" in rec or "inappropriate in the historic district" in rec or ("compatible" in rec and "historic district" in rec) or "incompatible with the historic district" in rec:
        return "Incompatible with Historic District"
    elif "duplicate submission" in rec:
        return "Duplicate Submission"
    elif "application has been inactive" in rec:
        return "Inactive Application"
    elif "installed prior" in rec or "prior to" in rec:
        return "Installed Prior to Review"
    else:
        return "Unknown"
```

```python
denial['recommendation_categorized'] = denial['recommendation'].apply(get_recommendation_categorized)
```
